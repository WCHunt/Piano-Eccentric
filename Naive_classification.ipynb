```python
import os
import mido

def classify_track(midi_file):
    # Separate notes into left and right hand tracks
    midi = mido.MidiFile(midi_file)
    left_hand_notes = []
    right_hand_notes = []
    for track in midi.tracks:
        if 'note' in track.name.lower():
            for msg in track:
                if msg.type == 'note_on':
                    # Determine which hand is playing this note
                    if msg.note < 60:
                        left_hand_notes.append(msg)
                    else:
                        right_hand_notes.append(msg)

    # Calculate note distribution for left hand
    left_hand_octave_counts = [0] * 9
    for note in left_hand_notes:
        octave = note.note // 12
        left_hand_octave_counts[octave] += 1

    # Calculate note distribution for right hand
    right_hand_octave_counts = [0] * 9
    for note in right_hand_notes:
        octave = note.note // 12
        right_hand_octave_counts[octave] += 1

    # Determine which hand is playing the track
    if sum(left_hand_octave_counts) > sum(right_hand_octave_counts):
        return "left hand"
    else:
        return "right hand"

# Directory containing MIDI files
midi_dir = '/content/drive/MyDrive/Midi_files/Parts'

# Loop through MIDI files in directory and classify each track
for file in os.listdir(midi_dir):
    if file.endswith('.mid'):
        midi_file = os.path.join(midi_dir, file)
        hand = classify_track(midi_file)
        print(f"{file}: {hand}")
```
